Carga el dataset de casas de California que está en la ruta "./data/california_cities.csv". Muestra parte de su contenido y su descripción. Deshazte de las filas con nulos.

```python
df_california=pd.read_csv("./data/california_cities.csv")
df_california.sample(10)
```

```python
df_california.info()
```

```python
df_california=df_california.dropna(axis="index")
df_california.info()
```

```python
top_poblacion=df_california.nlargest(5,"population_total")
top_poblacion
```

```python
outliers=df_california.nlargest(2,"population_total")
outliers
```

```python
coordenada_outlier_1=(float(outliers.iloc[0]["area_water_km2"]),float(outliers.iloc[0]["population_total"]/1000))
coordenada_outlier_1
```

```python
coordenada_outlier_2=(float(outliers.iloc[1]["area_water_km2"]),float(outliers.iloc[1]["population_total"]/1000))
coordenada_outlier_2
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
fig,ax=plt.subplots(2,2,figsize=(20,20))
colores= ["blue","grey","pink","olive","brown",]
bars=ax[0,0].bar(top_poblacion["city"],top_poblacion["population_total"]/1000, color=colores,alpha=0.7)
ax[0,0].set_ylim(0,5000)
ax[0,0].set_title("Ciudades más pobladas de California")
ax[0,0].set_xlabel("Ciudades")
ax[0,0].set_ylabel("Habitantes (miles)")
for barra in bars:
    altura_barra= barra.get_height()
    ax[0,0].text(barra.get_x()+barra.get_width()/2,altura_barra+100,f"{altura_barra}",ha="center")

ax[0,1].hist(df_california["area_total_km2"],bins=50,alpha=0.7)
font_dict={
    "size":18,
    "family":"Bradley Hand itc",
    "color":"black"
}
ax[0,1].set_xlabel("Superficie ($km^2$)",fontdict=font_dict)
ax[0,1].set_ylabel("Número de Ciudades",fontdict=font_dict)

ax[1,0].scatter(df_california["area_water_km2"],df_california["population_total"]/1000,color="green",alpha=0.6)
ax[1,0].set_title("Población por kilometro cuadrado de agua",size=13)
ax[1,0].set_xlabel("Área de agua en la ciudad ($km^2$)")
ax[1,0].set_ylabel("Población (miles habitantes)")
props_flecha={
    "arrowstyle":"->",
    "color":"black"}
ax[1,0].annotate("Outliers",xy =coordenada_outlier_1, xytext=(60, 3500),
             arrowprops=props_flecha,color="black")
ax[1,0].annotate("        ",xy =coordenada_outlier_2, xytext=(60, 3500),
             arrowprops=props_flecha,color="black")

ax[1,1].boxplot(df_california["elevation_m"],vert=False,patch_artist=True,boxprops={"facecolor":"red","color":"black"})
ax[1,1].set_title("Diagrama de Caja de la Altitud de las ciudades de California",size=13)
ax[1,1].set_xlabel("Altitud (m)")
```
